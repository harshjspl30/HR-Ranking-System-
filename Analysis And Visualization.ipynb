```python
#Encoding the dependented variable to binary > Joined to '0' Not Joined to '1

y.replace(['Joined','Not Joined'],[0,1],inplace=True)
```

```python
#From all above observation we can observe the target variable which is "Status" is inbalance in nature which is we can observe 
#from above plot number of candidates joined are more in number than the Not Joined once. Even though the model accuracy is good 
#enough but the predications made are wrong we knew it. So inorder the make dataset balance in nature we shall use the Synthetic 
#Minority Oversample TechniquE ( SMOTE ) for it to make it balance
```

```python
#Balancing the dataset using the SMOTE
```

```python
#Importing the imblearn

import imblearn
from imblearn.over_sampling import SMOTE #Importing SMOTE
```

```python
x_resam,y_resam = SMOTE(k_neighbors=5).fit_resample(xd,y)
```

```python
#Count plot
fig,ax=plt.subplots(figsize=(10,5))
sns.countplot(pd.Series(y_resam),ax=ax)
plt.show()
```

```python
#Random seed
np.random.seed(1001)
```

```python
#Importing train & test split
from sklearn.model_selection import train_test_split
```

```python
#Spliting to train,test
x_train,x_test,y_train,y_test = train_test_split(x_resam,y_resam,test_size=0.2)
```
